```python
import pandas as pd
from sklearn.model_selection import train_test_split
#df1=df.drop(labels=['respond_ind','user_id'], axis=1)
#Y=pd.Dataframe(df['respond_ind'])
df.fillna(value=-999,inplace=True)
Y=df['response_ind'].values
X=df.drop(labels=['response_ind','user_id'],axis=1, inplace=False)
```

```python
x_train,x_valid,y_train,y_valid=train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import make_scorer, accuracy_score 
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
# train with Gradient Boosting algorithm
# compute the accuracy scores on train and validation sets when training with different learning rates

learning_rates = [0.05, 0.1,0.25,0.5,0.75, 1]
#max_depth=[2,3,4,5,6]
for learning_rate in learning_rates:
    #for max_depth in max_depth:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features='auto', max_depth = 2, random_state = 0)
    gb.fit(x_train, y_train)
    print("Learning rate: ", learning_rate)
    #print("Max Depth: ",max_depth)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(x_valid, y_valid)))
    print()
```

```python
rf_clf = RandomForestClassifier()

parameters_rf = {"n_estimators": [4, 5, 6, 7, 8, 9, 10, 15], "criterion": ["gini", "entropy"], "max_features": ["auto", "sqrt", "log2"], 
                 "max_depth": [2, 3, 5, 10], "min_samples_split": [2, 3, 5, 10]}

grid_rf = GridSearchCV(rf_clf, parameters_rf, scoring=make_scorer(accuracy_score))
grid_rf.fit(x_train, y_train)

rf_clf = grid_rf.best_estimator_

rf_clf.fit(x_train, y_train)
pred_rf = rf_clf.predict(x_valid)
acc_rf = accuracy_score(y_valid, pred_rf)

print("The Score for Random Forest is: " + str(acc_rf))
```
